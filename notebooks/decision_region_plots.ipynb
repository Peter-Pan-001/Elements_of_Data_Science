```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')

from sklearn import datasets

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB

import sys
sys.path.insert(0,'../src/')

# taken from the mlxtend library
# https://rasbt.github.io/mlxtend/
# conda install mlxtend --channel conda-forge
from decision_regions import plot_decision_regions
```

```python
iris = datasets.load_iris()
```

```python
iris.target_names
```

```python
X = iris.data[:,[0,1]]
y = iris.target
y_setosa = (y == 0).astype(int)
y_virginica = (y == 2).astype(int)
```

```python
_ = plt.scatter(X[y==0][:,0], X[y==0][:,1],c='r',marker='^')
_ = plt.scatter(X[y==1][:,0], X[y==1][:,1],c='b',marker='s')
_ = plt.scatter(X[y==2][:,0], X[y==2][:,1],c='g',marker='o')
```

```python
knn = KNeighborsClassifier(n_neighbors=3).fit(X,y_setosa)

_ = plot_decision_regions(X,y_setosa,clf=knn)
```

```python
lr = LogisticRegression().fit(X,y_setosa)

_ = plot_decision_regions(X,y_setosa,clf=lr)
```

```python
lr = LogisticRegression().fit(X,y_virginica)

_ = plot_decision_regions(X,y_virginica,clf=lr)
```

```python
dt = DecisionTreeClassifier().fit(X,y_setosa)

_ = plot_decision_regions(X,y_setosa,clf=dt)
```

```python
svm = SVC().fit(X,y_setosa)

_ = plot_decision_regions(X,y_setosa,clf=svm)
```

```python
knn.fit(X,y_virginica)
lr.fit(X,y_virginica)
dt.fit(X,y_virginica)
svm.fit(X,y_virginica)

fig,ax = plt.subplots(2,2,figsize=(12,12))
_ = plot_decision_regions(X,y_virginica,clf=knn,ax=ax[0][0]);_ = ax[0][0].set_title('knn')
_ = plot_decision_regions(X,y_virginica,clf=lr, ax=ax[0][1]);_ = ax[0][1].set_title('lr')
_ = plot_decision_regions(X,y_virginica,clf=dt, ax=ax[1][0]);_ = ax[1][0].set_title('dt')
_ = plot_decision_regions(X,y_virginica,clf=svm,ax=ax[1][1]);_ = ax[1][1].set_title('svm')
```
